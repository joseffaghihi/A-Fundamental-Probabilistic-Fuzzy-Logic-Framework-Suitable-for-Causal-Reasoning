Let $\Omega\subseteq \mathbb{R}$ be a set, and $A$ be a fuzzy attribute of $\Omega$.

We define $\mathbb{P}(\Omega\text{ is } A)$ to be the chance that a randomly chosen element of $\Omega$ has the attribute $A$. Indeed, $\mathbb{P}(\Omega \text{ is } A)$ is obtained in two steps:

1. Randomly selecting an element $\omega$ from $\Omega$,

2. Doing a new random experiment for selected $\omega$ in Step (1) with the following outcomes:
\begin{equation*}
\{\text{$\omega$ is selected as $A$}, \text{Nothing is selected as $A$}\}.
\end{equation*}

We denote the probability value associated to "$\omega$ is selected as $A$" in Step (2) of the above experiment by $\mathbb{P}(\omega\text{ is }A)$. In Simple Fuzzy Model (see ***), we have that $\mathbb{P}(x\text{ is }A)=\mu_A(x)$ for any $x\in \Omega$, where $\mu_A$ is the membership function associated to $A$. 

In the following, we provide an example of this model. 


**Example**

The probabilities of reaching reproductive capability of a certain species in different days after the birth is given in Table 1. The bar chart of these probability values is provided in Figure 1. For simplicity, the small probability values for more than 199 days are ignored and assumed to be zero. We assume that $\Omega$ is the set of all integers from 0 to 199 reflecting the days order. Let us assume that $X$ is the inclusion random variable associated to $\Omega$. Consider the fuzzy attributes "early", "normal" and "late" for $\Omega$. Denote "early", "normal" and "late" by $A$, $B$ and $C$, respectively. Now, we define the fuzzy sets associated to $A$, $B$ and $C$ as follows:
	\begin{align*}
		\mu_A(x)&=\left\{\begin{array}{ll}
			1 & ,0\le x\le 40\\
			\dfrac{100-x}{60}& , 40\le x\le 100\\
			 0&,\text{otherwise}\end{array} \right.,\\
		\mu_B(x)&=\left\{\begin{array}{ll}
			\dfrac{x}{80}& , 0\le x\le 80
			\\
			1 &, 80\le x\le 120\\
			\dfrac{200-x}{80} &, 120\le x\le 200\\ 
			0&,\text{otherwise}\end{array} \right.,\\
		\mu_C(x)&=\left\{\begin{array}{ll}
			\dfrac{x-100}{60}&, 100\le x\le 160\\
			1 & ,160\le x\le 200\\
			 0&,\text{otherwise}\end{array} \right..\\
	\end{align*}

```python
import numpy as np
```

```python
# Calculating the probability of each single day 
#(i.e. the probability of the interval including the day divided by 10)

prob = np.zeros(200)

for i in range(0,10):
  prob[i] = 0

for i in range(10,20):
  prob[i] = 0.00015

for i in range(20,30):
  prob[i] = 0.0002

for i in range(30,40):
  prob[i] = 0.0005

for i in range(40,50):
  prob[i] = 0.002

for i in range(50,60):
  prob[i] = 0.005

for i in range(60,70):
  prob[i] = 0.0085

for i in range(70,80):
  prob[i] = 0.0105

for i in range(80,90):
  prob[i] = 0.013

for i in range(90,100):
  prob[i] = 0.0135

for i in range(100,110):
  prob[i] = 0.012

for i in range(110,120):
  prob[i] = 0.01

for i in range(120,130):
  prob[i] = 0.0075

for i in range(130,140):
  prob[i] = 0.006

for i in range(140,150):
  prob[i] = 0.004

for i in range(150,160):
  prob[i] = 0.003

for i in range(160,170):
  prob[i] = 0.002

for i in range(170,180):
  prob[i] = 0.0014

for i in range(180,190):
  prob[i] = 0.00065

for i in range(190,200):
  prob[i] = 0.0001
```

```python
# Defining fuzzy sets associating to "low", "medium", and $high".

def low(x):
  if 0 <= x <= 40:
    return 1
  elif 40 <= x <= 100:
    return (100-x)/60
  else:
    return 0

def med(x):
  if 0 <= x <= 80:
    return x/80
  elif 80 <= x <= 120:
    return 1
  elif 120 <= x <= 200:
    return (200-x)/80
  else:
    return 0

def high(x):
  if 100 <= x <= 160:
     return (x-100)/60
  elif 160 <= x <= 200:
    return 1
  else:
    return 0
```

```python
# Plotting the fuzzy sets associated to "early", "normal" and "late".
import matplotlib.pyplot as plt
# x axis values
x1 = [0,20,50]
x2 = [0, 40,60,100]
x3 = [50, 80,100]
# Corresponding y axis values
y1 = [1,1,0]
y2 = [0, 1, 1,0]
y3 = [0, 1,1]
 
# Plotting the points
plt.plot(x1, y1, label = "low")
plt.plot(x2, y2, label = "medium")
plt.plot(x3, y3, label = "high")
# Naming the x axis
plt.xlabel('Day- axis')
# Naming the y axis
plt.ylabel('Membership Degree')
plt.legend()
 
# Function to show the plot
plt.show()
```

Now, we calculate the following probability values:

\begin{align*}
		\mathbb{P}(\Omega\text{ is } A)&=\sum_{x\in \Omega}\mu_A(x)\mathbb{P}(X=x).\\
		\mathbb{P}(\Omega\text{ is } B)&=\sum_{x\in \Omega}\mu_B(x)\mathbb{P}(X=x).\\
		\mathbb{P}(\Omega\text{ is } C)&=\sum_{x\in \Omega}\mu_C(x)\mathbb{P}(X=x).
	\end{align*}

```python
prob_omega_is_low = 0

for i in range(0,200):
  prob_omega_is_low += low(i)*prob[i]

prob_omega_is_med = 0

for i in range(0,200):
  prob_omega_is_med += med(i)*prob[i]

prob_omega_is_high = 0

for i in range(0,200):
  prob_omega_is_high += high(i)*prob[i]
```

```python
print("P(Omega is low) = ", prob_omega_is_low, "\n")
print("P(Omega is med) = ", prob_omega_is_med, "\n")
print("P(Omega is high) = ", prob_omega_is_high, "\n")
```

Then, we calculate the following expected values (for more details about the following  random variables, see the paper):

\begin{align*}
		\mathbb{E}(\xi_{X,A})&=x_A+\sum_{x\in \Omega}(x-x_A)\mu_A(x)\mathbb{P}(X=x),\\
		\mathbb{E}(\xi_{X,B})&=x_B+\sum_{x\in \Omega}(x-x_B)\mu_B(x)\mathbb{P}(X=x), \\
		\mathbb{E}(\xi_{X,C})&=x_C+\sum_{x\in \Omega}(x-x_C)\mu_C(x)\mathbb{P}(X=x).
	\end{align*}
  To do so, we assume that $x_A=x_C=100$ and $x_B=200$, where these elements are base elements (i.e. $\mathbb{P}(x_A\text{ is } A)=0$, $\mathbb{P}(x_B\text{ is } B)=0$, and $\mathbb{P}(x_C\text{ is } C)=0$).

```python
ex_xi_Xlow = 100

for i in range(0,200):
  ex_xi_Xlow += low(i)*(i-100)*prob[i]

ex_xi_Xmed = 200

for i in range(0,200):
  ex_xi_Xmed += med(i)*(i-200)*prob[i]

ex_xi_Xhigh = 100

for i in range(0,200):
  ex_xi_Xhigh += high(i)*(i-100)*prob[i]
```

```python
print("E(xi_{X,low}) = ", ex_xi_Xlow, "\n")
print("E(xi_{X,med}) = ", ex_xi_Xmed, "\n")
print("E(xi_{X,high}) = ", ex_xi_Xhigh, "\n")
```
